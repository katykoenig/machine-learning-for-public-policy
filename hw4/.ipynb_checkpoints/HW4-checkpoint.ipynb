## Preprocess

We first load in the data and do some preprocessing. Note that we are analyzing the entire dataset (and not temporally split data).

```python
df = pl.read('projects_2012_2013.csv')
df.head()
```

```python
to_dummy = ['resource_type', 'grade_level', 'school_metro',
            'teacher_prefix', 'poverty_level']
to_discretize = ['school_charter', 'eligible_double_your_impact_match', 'school_magnet']
# Note that we drop many columns here only for ease of analysis below
# i.e. data that is too multidimensional would be difficult to concise analyze here on a cluster
to_drop = ['school_latitude', 'school_longitude', 'datefullyfunded', 'date_posted', 'primary_focus_subject', 'funded_by_deadline', 'primary_focus_area',
           'secondary_focus_subject', 'secondary_focus_area', 'school_city', 'school_city', 'school_state',
           'school_county', 'school_district', 'teacher_acctid', 'schoolid', 'school_ncesid']
```

```python
cleaned_df, kept, master_dict = pl.preprocess(df, to_dummy, to_discretize)
# gets output label for later analysis
updated = pl.check_for_funding(cleaned_df, timeframe=60)
#drops unnecessary columns (those that cannot logically be discretized or made into dummies)
processed = updated[updated.columns[~updated.columns.isin(to_drop)]]
y_col = updated['funded_by_deadline']
features = processed.columns
```

```python
processed.describe()
```

## Clustering Analysis on Data Overall

```python
output_df, model, labels, centroids = pl.make_kmeans_model(processed, 3)
```

Print summary stats for each cluster

```python
for cluster in set(labels):
    print("Cluster Number: " + str(cluster))
    print(pl.summary_stats(output_df[output_df['result'] == cluster]))
    print(pl.show_distribution(output_df[output_df['result'] == cluster]))
    print()
```

```python
# add y_labels back
output_df['funded_by_deadline'] = y_col
# Print Summary of Important Features
pl.describe_clusters(output_df, 'result', 'funded_by_deadline', features)
```

## Clustering Analysis on Top 5% of predicted projects least likely to be funded

Let's find the top 5% overall via a quick regression on the same data (with the same features) we used for clustering above

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import pandas as pd
```

```python
x_train, x_test, y_train, y_test = train_test_split(processed[features], y_col, test_size=0.3, random_state=57)
clf = LogisticRegression()
clf.fit(x_train, y_train)
predicted_scores = clf.predict_proba(x_test)[:, 1]
outcome = pd.DataFrame(dict(projectid = y_test.index, predicted_prob = predicted_scores, outcome_label=list(y_test)))
outcome = outcome.sort_values('predicted_prob', ascending=False)
top_five_pct = outcome.head(int(len(outcome)*(5/100)))
top_five_df = pd.merge(processed, top_five_pct[['projectid']], on='projectid')
y_labels = top_five_pct['outcome_label']
top_five_df = top_five_df.set_index('projectid')
```

```python
output_df2, model2, labels2, centroids2 = pl.make_kmeans_model(top_five_df, 3)
```

```python
for cluster in set(labels2):
    print("Cluster Number: " + str(cluster))
    print(pl.summary_stats(output_df2[output_df2['result'] == cluster]))
    print(pl.show_distribution(output_df2[output_df2['result'] == cluster]))
    print()
```

```python
# add y_labels back
output_df2['funded_by_deadline'] = list(y_labels)
# Print Summary of Important Features
pl.describe_clusters(output_df2, 'result', 'funded_by_deadline', features)
```
